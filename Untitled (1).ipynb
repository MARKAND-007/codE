```python
# ============================================
# 📊 MULTI-DATASET ANALYZER (Excel Based)
# ============================================

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 📥 Load Dataset
# -------------------------------
def load_data(file_path):
    try:
        df = pd.read_excel(file_path)
        df.columns = df.columns.str.strip()
        print("\n✅ Dataset Loaded Successfully!\n")
        print(df.head())
        return df
    except Exception as e:
        print("❌ Error loading file:", e)
        return None


# -------------------------------
# 📊 Basic Info
# -------------------------------
def basic_info(df):
    print("\n📌 BASIC INFORMATION\n")
    print(df.info())
    print("\n📊 Statistical Summary:\n")
    print(df.describe())


# -------------------------------
# 🔍 Missing Values
# -------------------------------
def check_missing(df):
    print("\n⚠️ Missing Values:\n")
    print(df.isnull().sum())


# -------------------------------
# 📈 Column Analysis
# -------------------------------
def column_analysis(df):
    print("\nAvailable Columns:\n", list(df.columns))
    col = input("\nEnter column name to analyze: ")

    if col in df.columns:
        print("\n📊 Value Counts:\n")
        print(df[col].value_counts())

        if pd.api.types.is_numeric_dtype(df[col]):
            print("\n📈 Column Stats:\n")
            print(df[col].describe())
    else:
        print("❌ Column not found!")


# -------------------------------
# 📊 Plot Graph
# -------------------------------
def plot_graph(df):
    print("\nAvailable Columns:\n", list(df.columns))
    col = input("\nEnter column name for graph: ")

    if col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].plot(kind='line')
            plt.title(f"{col} Trend")
            plt.xlabel("Index")
            plt.ylabel(col)
            plt.show()
        else:
            df[col].value_counts().plot(kind='bar')
            plt.title(f"{col} Distribution")
            plt.show()
    else:
        print("❌ Column not found!")


# -------------------------------
# 📊 Filter Data
# -------------------------------
def filter_data(df):
    print("\nAvailable Columns:\n", list(df.columns))
    col = input("\nEnter column name to filter: ")

    if col in df.columns:
        val = input("Enter value to filter: ")
        result = df[df[col].astype(str) == val]
        print("\n🔍 Filtered Data:\n")
        print(result)
    else:
        print("❌ Column not found!")


# -------------------------------
# 🧠 Menu System
# -------------------------------
def menu(df):
    while True:
        print("\n================================")
        print("📊 DATA ANALYSIS MENU")
        print("1. Basic Info")
        print("2. Missing Values")
        print("3. Column Analysis")
        print("4. Plot Graph")
        print("5. Filter Data")
        print("6. Exit")
        print("================================")

        choice = input("Enter choice: ")

        if choice == '1':
            basic_info(df)
        elif choice == '2':
            check_missing(df)
        elif choice == '3':
            column_analysis(df)
        elif choice == '4':
            plot_graph(df)
        elif choice == '5':
            filter_data(df)
        elif choice == '6':
            print("👋 Exiting...")
            break
        else:
            print("❌ Invalid choice!")


# -------------------------------
# 🚀 Main Program
# -------------------------------
if __name__ == "__main__":
    file_path = input("📂 Enter Excel file path: ")
    df = load_data(file_path)

    if df is not None:
        menu(df)
```
